# Explore songs that are in the popularity 100

- there is 1 song that is popular and contains the explicit word, namely greedy, and has been played 30 times
- there are 2 popular and non-explicit songs, namely Cruel Summer played 18 times, End of Beginning 2 times

```python
# songs that are popular and contain explicit words

true_explicit =df_id[(df_id.is_explicit==True) & (df_id.popularity==100)]
print('jumlah lagu yang populer dan eksplicit\n', true_explicit.Songs.value_counts(), '\n')
print(true_explicit[['snapshot_date','Songs']])


# popular and non-explicit songs

false_explicit = df_id[(df_id.is_explicit==False) & (df_id.popularity==100)]
print('\njumlah lagu yang populer dan non-eksplicit\n', false_explicit.Songs.value_counts())
print('\n',false_explicit[['snapshot_date','Songs']])
```

```python
# genre yang dimiliki oleh lagu yang mendapatkan kepopuleran 100 
genre_popularity_100 = df_id[df_id.popularity==100]
for song in genre_popularity_100['Songs'].unique():
    display(genre_popularity_100[genre_popularity_100['Songs'] == song].head(1)[['artists','Songs','danceability', 'energy', 'valence']])
    
```

.

On average, songs that are in the highest popularity (100) are more dominant towards energetic songs. Usually energetic tracks feel fast, loud and noisy. For example, death metal has high energy and valence, namely the positivity the music conveys, such as happiness or cheerfulness.

```python
# average popular song per genre
mean = genre_popularity_100.groupby('artists').agg(
{
    'danceability': 'mean',
    'energy' : 'mean',
    'valence' : 'mean',
    'acousticness' : 'mean'
})
print(mean)

mean.mean()
```

## Genre : danceability, energy and valence selama di indonesia

In the last few months there has been an increase in song playback with a level of danceability (mid of month 3 - end of month 4) followed by a decrease in energy level, there were 4 times when there was an increase in danceability and the highest value of danceability occurred on 11-4-2024:

```python
time_series =df_id.groupby(['snapshot_date']).agg({
    'danceability': 'mean',
    'energy' : 'mean',
    'valence' : 'mean',
})
id_time_series = time_series.sort_index()
#print(id_time_series.head())

id_time_series.plot(
    y =['danceability', 'energy', 'valence'],
    color = ['blue', 'green', 'red'],
linestyle= '-', marker= 'o',
figsize =(30,20),
title = 'danceability, energy and valence over time in the indonesia',
xlabel = 'snapshotDate', ylabel='value')

plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```
