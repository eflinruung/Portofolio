# daily_rank 1

```python
daily_rank_1= df_id[df_id['daily_rank'] == 1]

songs_daily_rank_1 = df_id.groupby(['snapshot_date','name','album_release_date', daily_rank_1['daily_rank']]).size().reset_index().rename(columns={0:'jumlah'})
# songs_daily_rank_1.sort_values(by ='snapshot_date', ascending=False).head(26)

```

```python
groupby_songs_daily_rank_1 =songs_daily_rank_1.groupby('name').size().reset_index().rename(columns={0:'Amount'}).sort_values('Amount',ascending=False)
display(groupby_songs_daily_rank_1)

def lagu_daily_rank_1(song):
    music = daily_rank_1[(daily_rank_1['name'] == song)]
    print('album_release_date',music.album_release_date.head(1),'\n')
    print(music.snapshot_date.unique())
    
    
lagu_daily_rank_1('penjaga hati')
```

The song Penjaga Hati appeared with the highest daily ranking of 1, namely for 145 days

The song Penjaga Hati was released on 06-21-2023 but can still get daily ranking 1 in 2024

sequentially starting from the date the data was taken on 10-18-2023 to 03-13-2024 
except on 03-05-2024 and 03-06-2024 which did not get a daily_rank of 1

```python
# genre belonging to the song that gets 100 popularity
genre_popularity_100 = df_id[df_id.popularity==100]
for song in genre_popularity_100['name'].unique():
    display(genre_popularity_100[genre_popularity_100['name'] == song].head(1)[['artists','name','danceability', 'energy', 'valence']])
    
```

```python
# average popular song genre
mean = genre_popularity_100.groupby('artists').agg(
{
    'danceability': 'mean',
    'energy' : 'mean',
    'valence' : 'mean',
    'acousticness' : 'mean'
})
print(mean)

mean.mean()
```

on average, songs that are included in the highest popularity (100) are more dominant
towards energetic songs. Usually energetic tracks feel fast, loud and noisy. For example, death metal has high energy and, valence namely the positivity the music conveys, such as happiness or cheerfulness.

```python
time_series =df_id.groupby(['snapshot_date']).agg({
    'danceability': 'mean',
    'energy' : 'mean',
    'valence' : 'mean',
})
id_time_series = time_series.sort_index()
print(id_time_series.head())

id_time_series.plot(
    y =['danceability', 'energy', 'valence'],
    color = ['blue', 'green', 'red'],
linestyle= '-', marker= 'o',
figsize =(30,20),
title = 'danceability, energy and tempo over time in the indonesia',
xlabel = 'snapshotDate', ylabel='value')

plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```
